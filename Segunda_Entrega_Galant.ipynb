```python
# Se hace la primera pregunta
prompt = "Cuales son las principales causas del calentamiento global."

# Crear la solicitud para obtener la respuesta de IA usando Groq
respuesta = cliente.chat.completions.create(
    model = "llama3-8b-8192",
    messages = [
        {"role": "user", "content": prompt}],
    max_tokens = 500
)

# Mostrar la respuesta del modelo
print(respuesta.choices[0].message.content)
```

```python
# Primer ejemplo (Few-shot)
# Realizamos otra pregunta como lo hicimos en el ejemplo original
prompt_2 = "¿Cómo afecta el cambio climático a los ecosistemas acuáticos?"

respuesta_2 = cliente.chat.completions.create(
    model = "llama3-8b-8192",
    messages = [

        {"role": "system", "content": "Eres un asistente util"},
        {"role": "user", "content": prompt},
        {"role": "assistant", "content": respuesta.choices[0].message.content},
        {"role": "user", "content": prompt_2}],
    max_tokens = 250
)

print(respuesta_2.choices[0].message.content)
```

```python
# Pregunta mas especifica
prompt_3 = "- ¿Qué acciones individuales pueden tomar los estudiantes para reducir su huella carbonica diaria?."

respuesta_3 = cliente.chat.completions.create(
    model = "llama3-8b-8192",
    messages = [

        {"role": "system", "content": "Eres un asistente util"},
        {"role": "user", "content": prompt_2},
        {"role": "user", "content": prompt_3}],
    max_tokens = 250
)

print(respuesta_3.choices[0].message.content)
```

```python
# Pregunta final
prompt_4 = "¿cómo podrían estos hábitos sostenibles influir en políticas públicas más amplias si fueran adoptados por comunidades enteras?"

respuesta_final = cliente.chat.completions.create(
    model = "llama3-8b-8192",
    messages = [
        {"role": "user", "content": prompt_4}],
    max_tokens = 300
)

print(respuesta_final.choices[0].message.content)
```
